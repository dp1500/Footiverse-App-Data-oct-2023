```python
# import pandas as pd


# # # Load data
# players_df = pd.read_csv("/content/male_players.csv")
# teams_df = pd.read_csv("/content/male_teams.csv")

import json

# Convert players and teams CSV to JSON
players_df.to_json("players.json", orient="records")
teams_df.to_json("clubs.json", orient="records")

# Create Tournament JSON
unique_countries = teams_df['nationality_name'].unique()
tournament_data = []

for i, country in enumerate(unique_countries):
    if pd.notna(country):
        tournament_data.append({
            'tournament_id': i,
            'name': f"{country} Cup",
            'type': "Country Cup",
            'country_id': i
        })

with open('tournament.json', 'w') as file:
    json.dump(tournament_data, file)

# Create League JSON
unique_leagues = teams_df[['league_id', 'league_name', 'league_level', 'nationality_id']].drop_duplicates()
league_data = unique_leagues.to_dict(orient="records")

with open('league.json', 'w') as file:
    json.dump(league_data, file)

# Create Enhanced League JSON for matchday tracking
teams_performance_data = []

for index, row in teams_df.iterrows():
    for matchday in range(1, 39):  # Assuming a league has 38 matchdays
        teams_performance_data.append({
            'team_id': row['team_id'],
            'team_name': row['team_name'],
            'matchday': matchday,
            'points_scored': 0,
            'goals_scored': 0,
            'goals_taken': 0,
            'goal_difference': 0
        })

with open('teams_performance.json', 'w') as file:
    json.dump(teams_performance_data, file)
```
